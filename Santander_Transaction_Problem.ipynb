```python
#Load data
train_org = pd.read_csv("train.csv")
test_org = pd.read_csv("test.csv")
```

```python
train_org.head()
```

```python
#Considering Only Continuos Variables in DataSets
train_org1 = train_org.drop(['target', 'ID_code'], axis = 1) 
test_org1 = train_org.drop(['ID_code'], axis = 1) 
```

```python
#Missing Values Analysis
missing_val = pd.DataFrame(train_org.isnull().sum())
missing_val
```

```python
#Import Libraries for decision tree
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#Replace target categories with Yes or No
train_org['target'] = train_org['target'].replace(0, 'No')
train_org['target'] = train_org['target'].replace(1, 'Yes')
```

```python
train_org.head(10)
```

```python
#Divide data into train and test
X = train_org.values[:, 2:202]
Y = train_org.values[:,1]

X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.25)
```

```python
#Decision Tree
C50_model = tree.DecisionTreeClassifier(criterion='entropy').fit(X_train, y_train)
```

```python
#predict new test cases
C50_Predictions = C50_model.predict(X_test)
```

```python
#Create dot file to visualise tree  #http://webgraphviz.com/
dotfile = open("pt.dot", 'w')
df = tree.export_graphviz(C50_model, out_file=dotfile, feature_names = train_org1.columns)
```

```python
#build confusion matrix
from sklearn.metrics import confusion_matrix 
CM = confusion_matrix(y_test, C50_Predictions)
CM = pd.crosstab(y_test, C50_Predictions)
```

```python
#let us save TP, TN, FP, FN
TN = CM.iloc[0,0]
FN = CM.iloc[1,0]
TP = CM.iloc[1,1]
FP = CM.iloc[0,1]
```

```python
#check accuracy of model
accuracy_score(y_test, C50_Predictions)*100
```

```python
#Accuracy Formula
((TP+TN)*100)/(TP+TN+FP+FN)
```

```python
#False Negative Rate
(FN*100)/(FN+TP)
```

```python
#Recall
(TP*100)/(FN+TP)
```

```python
#Precision
(TN*100)/(TN+FP)
```

```python
#select column indexes for independent variables
train_cols = train_org.columns[2:202]
```

```python
#Built Logistic Regression
import statsmodels.api as sm

logit = sm.Logit(train_org['target'], train_org[train_cols]).fit()

logit.summary()
```

```python
#Predict test data
test_org1['Actual_prob'] = logit.predict(test_org1[train_cols])
test_org1['ActualVal'] = 1
test_org1.loc[test_org1.Actual_prob < 0.5, 'ActualVal'] = 0
test_org1.head()
```

```python
#Build confusion matrix
CM2 = pd.crosstab(test_org1['target'], test_org1['ActualVal'])
CM2
```

```python
#let us save TP, TN, FP, FN
TN2 = CM2.iloc[0,0]
FN2 = CM2.iloc[1,0]
TP2 = CM2.iloc[1,1]
FP2 = CM2.iloc[0,1]
```
